```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}

class Card:
    def __init__(self,rank,suit):
        self.rank=rank
        self.suit=suit
        self.value=values[rank]
        
    def __str__(self):
        return f"{self.rank} of {self.suit}"

class Deck:
    def __init__(self):
        self.deck=[]
        for suit in suits:
            for rank in ranks:
                tempcard=Card(rank,suit)
                self.deck.append(tempcard)
    def shuffle(self):
        random.shuffle(self.deck)
    def deal_one(self):
        self.deck.pop(0)
            
class Hand:
    def __init__(self):
        self.cards=[]
        self.value=0
        
    def addone(self,card):
        self.cards.append(card)
        self.value+=values[card.rank]
        
class Chips:
    def __init__(self):
        self.tot=100
        self.bet=0
        
    def deb(self):
        self.tot-=self.bet
    def cred(self):
        self.tot+=2*(self.bet)
        
        
        
def hit(deck,hand):
    mycard=deck.deck.pop(0)
    hand.addone(mycard)
    print(mycard)
    
def take_bet(chips):
    
    while True:
        bet=int(input("Enter bet value"))
        if bet>chips.tot:
            print("Not enough chips")
        else:
            chips.bet=bet
            break
        
    

def hit_or_stand(deck,hand,chips):
    global playing # to control an upcoming while loop
    
    while True:
        what_next=input("hit or stand")
        if what_next=="hit":
            hit(deck,hand)
            if hand.value>21:
                player_busts(chips)
            
        if what_next=="stand":
            #laying =False
            break
    
    
def show_some(player,dealer):
    for card in player.cards:
        print(card)
        
    print(dealer.cards[0])
    
   
    
def show_all(player,dealer):
    for card in player.cards:
        print(card)
        
    for ard in dealer.cards:
        print(ard)
        
    print(player.value)
    print(dealer.value)
    
        
def player_busts(chips):
    print("PLayer bust")
    chips.deb()
      

def player_wins(chips):
    print("PLayer wins")
    chips.cred()
    

def dealer_busts(chips):
    print("Dealer Bust")
    chips.cred()
   
    
def dealer_wins(chips):
    print("Dealer wins")
    chips.deb()
    
    
def push():
    print("Tie")
    
    


    

while True:
    
    allcards=Deck()
    allcards.shuffle()
    player=Hand()
    dealer=Hand()
    
    player.addone(allcards.deck.pop(0))
    player.addone(allcards.deck.pop(0))
    
    
    dealer.addone(allcards.deck.pop(0))
    dealer.addone(allcards.deck.pop(0))
    p_chips=Chips()
    take_bet(p_chips)
    show_some(player,dealer)
    
    #hile playing=True:
    hit_or_stand(allcards,player,p_chips)
        
    if player.value<=21:
        show_all(player,dealer)
        if dealer.value<17:
            while dealer.value<17:    
                hit(allcards,dealer)
        if 17<=dealer.value<=21:
            if player.value>dealer.value:
                player_wins(p_chips)
            elif player.value<dealer.value:
                dealer_wins(p_chips)
                
            else:
                push()
                
        if dealer.value>21:
            dealer_busts(p_chips)
    
    print(p_chips.tot)
    nexround=input("Next Round?(Y/N)")
    if nexround=="Y":
        pass
    if nexround=="N":
        break
```
